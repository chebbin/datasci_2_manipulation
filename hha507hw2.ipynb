```python
import pandas as pd
import re
import numpy as np
```

# Data Manipulation in Python: Pandas, Polars, Dask and Modin

```python
df = pd.read_csv ('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df.tail(10)
```

```python
## Check for missing column values
print(df.isnull().sum())
```

```python
## Drop rows with any missing values
print(df.dropna(inplace=True))
```

```python
## Check for any duplicate rows and columns
print(df.drop_duplicates(inplace=True))
```

```python
## Remove leading and trailing white space from all columns
def clean_column_names(df):
  def clean_name(name):
      cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
      return cleaned_name.lower()
  df.columns = [clean_name(col) for col in df.columns]
  return df

df2 = clean_column_names(df)

print(df2.head(10))
```

# Changes to column names

The column names have now been changed to all lowercase letters, no white space and no special characters. The new names are listed below.

```python
list(df2)
```

# Remove Missing Values

```python
df2.replace('missing', pd.np.nan, inplace=True)
print(df2.head(10))
```

# Change String values to Integers


```python
df2['averageheartrate'] = df2['averageheartrate'].astype(float)
```

# Group by gender and average heart rate


```python
gp2 = df2.groupby('gender').agg({
    'averageheartrate': 'mean'
    })
gp2
```
